## Veriseti Ön İşleme

```python
X_train.info()
```

```python
## encoder :  sayısal değerlere döndürür
## SimpleImputer : eksik değerleri doldurma işini görüyor

from sklearn.preprocessing import OrdinalEncoder 
from sklearn.impute import SimpleImputer
```

```python
categorical_column = X_train.select_dtypes(include=["object"]).columns
numerical_column = X_train.select_dtypes(include=["int64" , "float64"]).columns
```

```python
print(categorical_column)

numerical_column
```

```python
# Numerical
numerical_imputer = SimpleImputer(strategy="median")

#Categorical
encoder = OrdinalEncoder(handle_unknown= "use_encoded_value", 
                         unknown_value= -1)
categorical_imputer = SimpleImputer(strategy="most_frequent")
```

```python
X_train[numerical_column]
X_train[numerical_column] = numerical_imputer.fit_transform(X_train[numerical_column])
X_train[numerical_column].isna().sum().sum()
```

```python
# Test datasında fit_transform yapmıyoruz, transorm yapıyoruz.

X_test[numerical_column] = numerical_imputer.transform(X_test[numerical_column])
```

```python
X_train
```

```python
X_train[categorical_column] = categorical_imputer.fit_transform(X_train[categorical_column])

X_test[categorical_column] = categorical_imputer.transform(X_test[categorical_column])
```

```python
X_train
```

```python
X_train[categorical_column] = encoder.fit_transform(X_train[categorical_column])
```

```python
X_test[categorical_column] = encoder.transform(X_test[categorical_column])
```

```python
X_train[categorical_column]
```

```python
X_train
```

## Modelin kurulması :

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
tahmın = model.predict(X_test)
tahmın
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, tahmın , squared=False)
```
